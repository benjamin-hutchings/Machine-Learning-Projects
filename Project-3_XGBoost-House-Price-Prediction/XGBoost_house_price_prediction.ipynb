```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
house_price_dataset
```

```python
house_price_df = pd.DataFrame(house_price_dataset['data'], columns=house_price_dataset['feature_names'])
```

```python
# add the target column (house price)
house_price_df['Price'] = house_price_dataset.target
```

```python
X = house_price_df.drop(['Price'], axis=1)
Y = house_price_df['Price']
```

```python
X
```

```python
Y
```

Split into training and test data

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training - XGBoost Regressor

```python
model = XGBRegressor()
```

```python
model.fit(X_train, y_train)
```

Evaluation on Training Data

```python
# Prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(y_train, training_data_prediction)
score_1
```

```python
# Mean absolute error
score_2 = metrics.mean_absolute_error(y_train, training_data_prediction)
score_2
```

Visualise

```python
plt.scatter(y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```

Test Data

```python
# Prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
score_1 = metrics.r2_score(y_test, test_data_prediction)
score_1
```

```python
score_2 = metrics.mean_absolute_error(y_test, test_data_prediction)
score_2
```
